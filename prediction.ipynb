```python
from pathlib import Path
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_path = Path.cwd() / "distilbert-finetuned-emotion" / "checkpoint-500"

print(model_path)
print(model_path.exists())

tokenizer = AutoTokenizer.from_pretrained(
    str(model_path),
    local_files_only=True
)

model = AutoModelForSequenceClassification.from_pretrained(
    str(model_path),
    local_files_only=True
)

print("Loaded successfully")
```

```python
import torch

text = "I am very happy today!"

inputs = tokenizer(
    text,
    return_tensors="pt",
    truncation=True,
    padding=True
)

with torch.no_grad():
    outputs = model(**inputs)

predicted_class = torch.argmax(outputs.logits, dim=1).item()

print("Predicted label:", predicted_class)
```

```python
classes = {0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4: 'fear', 5: 'surprise'}
```

```python
emotion = classes[predicted_class]
print("Emotion:", emotion)
```

```python
from transformers import pipeline, AutoConfig

# Define your mapping
id2label = {0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4: 'fear', 5: 'surprise'}
label2id = {v: k for k, v in id2label.items()}
config = AutoConfig.from_pretrained(str(model_path))
config.id2label = id2label
config.label2id = label2id

# Re-save the config to your folder
config.save_pretrained(model_path)


# Create a classifier pipeline using your local model
classifier = pipeline("text-classification", model=str(model_path), tokenizer=str(model_path))

result = classifier("I love this model!")
print(label2id)
print(result)
```
